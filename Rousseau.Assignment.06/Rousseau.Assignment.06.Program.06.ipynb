```python
import pandas as pd
```

1. Read the data from the CSV file into a DataFrame and display the first five rows.


```python
exam_data = pd.read_csv('exams.csv')
exam_data.head(5)
```

2. Use the pandas plot() method to create a box plot.

```python
exam_data.plot.box()
```

3. Use the pandas plot() method to create a plot like the one that follows.


```python
exam_data.plot.scatter(x='math score', y='reading score')
```

4. Group the data by the gender column and calculate the average scores. Then create a bar plot.

```python
exam_data['mean score'] = \
    (exam_data['reading score'] + exam_data['math score']  + exam_data['writing score'])/3

(exam_data.groupby('gender')['mean score'].mean()).plot.bar(title='Average Score by Gender')
```
