```python
import pandas as pd
```

1. Read the data from the .csv file into a DataFrame and display the first five rows.


```python
avocado_data = pd.read_csv('avocado.csv')
avocado_data.head(5)
```

2. Create a new DataFrame that contains the total of the Small Bags, Large Bags, and XLarge Bags
columns grouped by type, and then display the DataFrame.

```python
grouped_data = avocado_data.groupby('type').agg({'Small Bags': 'sum', 'Large Bags': 'sum', 'XLarge Bags': 'sum'}).reset_index()

# Create a DataFrame from the grouped data
grouped_avocado = pd.DataFrame(grouped_data)

# Print the DataFrame to see the result
print(grouped_avocado)
```

3. Use the grouped data to create a bar plot that shows the number of small, large, and extra-large bags
for both types of avocado.

```python
(grouped_avocado.set_index('type')).plot.bar(title='Number of Bags by Avocado Type')
```

4. Use the original data to create a scatter plot for the Total Volume and AveragePrice columns.


```python
avocado_data.plot.scatter(y='Total Volume', x='AveragePrice', title='Total Volume by Average Price of Avocado')
```
