1. Read the data from the CSV file into a DataFrame and display the first five rows

```python
# 1
ramen_data = pd.read_csv('ramen-ratings.csv')
ramen_data.head(5)
```

```python
# 2
sns.barplot(data=ramen_data, x='Style', y='Stars')
```

3. Create a bar plot to show the Stars by Style with a Seaborn generic method.

```python
# 3
sns.catplot(data=ramen_data, x='Style', y='Stars', kind='bar')
```

4. Modify the plot you just created to use a custom title and x-axis label.

```python
# 4
ramen_bar_graph = sns.catplot(data=ramen_data, x='Style', y='Stars', kind='bar')
for ax in ramen_bar_graph.axes.flat:
    ax.set(title='Stars by Ramen Style', xlabel='Ramen Style')
```

5. Use the query() method to filter the data to contain only data for Japan, India, Taiwan, and the U.S. Plot
the stars by style for each country on a subplot with two subplots per row.

```python
# 5
ramen_country_graph = sns.catplot(data=ramen_data.query('Country == "Japan" or Country == "Taiwan" or Country == "India"  or Country == "USA"'), 
    x='Style', y='Stars', kind='bar', col='Country', col_wrap=2, height=4)
ramen_country_graph.figure.suptitle('Ramen Stars by Country')

for i, ax in enumerate(ramen_country_graph.axes.flat): 
        ax.set(ylabel='Ramen Stars')
```

```python
# 6
ramen_country_graph = sns.catplot(data=ramen_data.query('Country == "Japan" or Country == "Taiwan" or Country == "India"  or Country == "USA"'), 
    x='Style', y='Stars', kind='bar', col='Country', col_wrap=2, height=4)
ramen_country_graph.figure.suptitle('Ramen Stars by Country')

countries = ['Japan', 'Taiwan', 'India', 'USA']

for i, ax in enumerate(ramen_country_graph.axes.flat): 
        ax.set(ylabel='Ramen Stars')
        ax.set_title(countries[i])
```
