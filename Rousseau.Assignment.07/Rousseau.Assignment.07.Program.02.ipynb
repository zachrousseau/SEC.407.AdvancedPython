```python
# 3
sns.relplot(data=avocado_data, x='AveragePrice', y='Total Volume', kind='scatter')
```

4. Create the same plot but with the hue parameter set to year and the dots for the total volume ranging in
size from 10 to 100.

```python
# 3
avocado_graph = sns.relplot(data=avocado_data, x='AveragePrice', y='Total Volume', kind='scatter', hue='year')

for ax in avocado_graph.axes.flat:
    ax.set(
        ylim=(10, 100)
    )
```
