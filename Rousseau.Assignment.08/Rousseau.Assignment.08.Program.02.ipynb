4. Write a function that takes a row as input and returns a shortened version of the values in the Variety
column. The values should consist of the first two words of the variety name, followed by a space and
an ellipsis (…). Hint: You’ll need to split the values into a list of words, get the first two words in the
list, add the ellipsis to the list, and then join the words in the list.

```python
def shorten_variety(row):

    variety_split = row.Variety.split(' ')

    if(len(variety_split) > 1):
        output = variety_split[0] + ' ' + variety_split[1] + ' ...'
    else:
        output = variety_split[0] + ' ...'

    return output
```

5. Apply the function you just wrote to the DataFrame, and store the results in a column named
ShortVariety. Then, display the first five rows to be sure this worked.

```python
for index, row in ramen_data.iterrows():
    ShortVariety = shorten_variety(row)
    
    ramen_data.loc[index, 'ShortVariety'] = ShortVariety
    # print(ShortVariety)

ramen_data.head(5)
```
