```python
import pandas as pd
import seaborn as sns
```

1. Read the data from the CSV file into a DataFrame called gold

```python
gold = pd.read_csv('gold.csv')
```

2. Run the info() method, and note the data types of the two columns

```python
gold.info()
```

3. Convert the Date column to the DateTime data type. Then, display the first five rows of data, and note
that the dates are for the first of each month starting in 1950.

```python
gold.Date = pd.to_datetime(gold.Date, format ='%Y-%m')
gold.head(5)
```

4. Plot the data with a Pandas line plot

```python
gold.plot.line(x='Date', y='Price')
```

5. Index the data on the Date column. Then, generate time periods for the first day of each year from 1950
to 2020.

```python
gold.set_index('Date', inplace=True)
yearly = pd.date_range('01/01/1950', '01/01/2020', freq='YS')
```

6. Reindex the data using the new time periods, and display the first five rows again to see that there is just
one row for each year.

```python
gold.reindex(yearly).head(5)
```

7. Plot the reindexed data with a Pandas line plot.

```python
gold.reindex(yearly).plot.line(y='Price')
```

8. Resample the data to a quarterly frequency and plot the data again

```python
gold.reset_index()
quarterly = pd.date_range('01/01/1950', '01/01/2020', freq='QS')
gold.reindex(quarterly).plot.line(y='Price')
```

9. Resample the data to a yearly frequency and plot the data one more time.

```python
gold.reset_index()
gold.reindex(yearly).plot.line(y='Price')
```

10. Plot the rolling mean for the data using a Pandas line plot. Experiment with different values for the
window parameter.

```python
gold.rolling(window=50).mean().plot.line(y='Price')
```
